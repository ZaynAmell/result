```python
splices=16
spl_order=2
channel_width=w[1].container.n_channels_in_file
domin=channel_width/splices
r=[]
for i in range(splices):
    x=w[1].get_freqs()[int(i*domin):int((i+1)*domin)]
    integrated_channel=w[1].get_power()[1][int(i*domin):int((i+1)*domin)]
    knots = np.arange(x[0], x[-1], (x[-1]-x[0])//spl_order+1)
    spl = interpolate.splrep(x, integrated_channel,t=knots[1:])
    chan_fit = interpolate.splev(x, spl)
    if i==0:
        r.append(chan_fit)
    else:
        r[0]=np.append(r[0],chan_fit)
w[1].plot_spectrum()
plt.plot(w[1].get_freqs(),r[0])
```
